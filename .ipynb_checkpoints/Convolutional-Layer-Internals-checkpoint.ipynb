```python
from scipy import misc
```

```python
img = misc.ascent()
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img,cmap="gray")
```

```python
filtera=numpy.array([     #Famous filter
    [1,2,1],
    [0,0,0],
    [-1,-2,-1]
]
)
```

```python
filtera.shape
```

```python
from scipy.signal import convolve2d 
```

```python
cimage=convolve2d(img,filtera)
```

```python
cimage
```

```python
cimage.shape
```

```python
plt.imshow(cimage,cmap="gray")
```

```python
cimage.shape
```

```python
#https://en.wikipedia.org/wiki/Kernel_(image_processing)
filterb=numpy.array([
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
]
)
```

```python
cimage=convolve2d(img,filterb)
```

```python
plt.imshow(cimage,cmap='gray')
```

```python
filterc = numpy.array([
    [7,38,55],
    [-32,-4,-8],
    [-55,-56,-58]
])
```

```python
cimage=convolve2d(img,filterc)
```

```python
plt.imshow(cimage,cmap="gray")
```

```python
cimage.shape
```

### Using custom local image

```python
local_image=cv2.imread("/root/Test-Images/test_image1.jpg")
```

```python
type(local_image)
```

```python
cv2.imshow("ORIGINAL_IMAGE", local_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
grey_local_image = cv2.cvtColor(local_image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("GRAYSCALE_IMAGE", grey_local_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
filtera=numpy.array([     #Famous filter
    [1,2,1],
    [0,0,0],
    [-1,-2,-1]
]
)
grey_local_image.shape
```

```python
cimg = convolve2d(grey_local_image,filtera)
```
